```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

URL = "http://localhost:8000/appointments-with-id-display-name/?skip=0&limit=100"
resp = requests.get(URL, json=True)

appointments_json = resp.json()
```

```python
appointments_df = pd.DataFrame.from_dict(appointments_json)
```

```python
appointments_df.groupby("date_and_time").agg({'patient_id': "count"})
```

```python
appointments_by_doctor = appointments_df.groupby("doctor_display_name").agg({'patient_id': "count"})
appointments_by_doctor
```

```python
appt_by_doctor_by_time = appointments_df.groupby(["doctor_display_name", "date_and_time"]).agg({'patient_id': "count"}).reset_index()
appt_by_doctor_by_time.head(5)
```

```python
availability_pivot = appt_by_doctor_by_time.pivot_table('patient_id', ['doctor_display_name'], 'date_and_time')
availability_pivot.head(5)
```

```python
available_schedules = availability_pivot.reset_index()
```

```python
schedules = available_schedules.replace(np.nan, 'Available')
schedules = schedules.replace(1.0, 'Busy')
schedules
```

```python
from IPython.display import JSON
import json
JSON(json.loads(schedules.to_json()), expanded=True)
```

```python
schedules_T = schedules.transpose()
schedules_T
```

```python
appt_by_doctor = appointments_by_doctor.reset_index()
doctors = appt_by_doctor['doctor_display_name']
counts = appt_by_doctor['patient_id']

# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))

# Horizontal Bar Plot
ax.barh(doctors, counts)
ax.vlines(x=16, ymin=-1, ymax=len(doctors), colors='r', linestyle='--', label="Max. appointments per day")

# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('Appointments by doctor',
             loc ='left', )
 

 
# Show Plot
plt.legend(loc='upper left')
# plt.show()
plt.savefig("sample.png", format="png", bbox_inches="tight")

```
