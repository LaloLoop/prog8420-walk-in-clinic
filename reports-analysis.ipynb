```python
import requests
import pandas as pd
import numpy as np
```

```python

URL = "http://localhost:8000/appointments-with-id-display-name/?skip=0&limit=100"
resp = requests.get(URL, json=True)

appointments_json = resp.json()
```

```python
appointments_df = pd.DataFrame.from_dict(appointments_json)
```

```python
appointments_df.head(10)
```

```python
appointments_df.groupby("prescription_display_name").agg({'patient_id': 'count'}).head(5)
```

```python
appointments_df.groupby("doctor_display_name").agg({'patient_id': "count"})
```

```python
appointments_df.groupby("staff_display_name").agg({'patient_id': "count"})
```

```python
appointments_df.groupby("date_and_time").agg({'patient_id': "count"})
```

```python
appointments_df.groupby("doctor_display_name").agg({'patient_id': "count"})
```

```python
appt_by_doctor_by_time = appointments_df.groupby(["doctor_display_name", "date_and_time"]).agg({'patient_id': "count"}).reset_index()
appt_by_doctor_by_time.head(5)
```

```python
availability_pivot = appt_by_doctor_by_time.pivot_table('patient_id', ['doctor_display_name'], 'date_and_time')
availability_pivot.head(5)
```

```python
available_schedules = availability_pivot.reset_index()
```

```python
schedules = available_schedules.replace(np.nan, 'Available')
schedules = schedules.replace(1.0, 'Busy')
schedules
```
